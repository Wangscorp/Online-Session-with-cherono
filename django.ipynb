```python
names=["Edwin","Mercy", "Cherono","Jane"]

print(names)
print(len(names))

names[-1]="Felista"
print(names)

# Append
names.append("Baraka")
print(names)

names[2:4]="Victor","Dan"
print(names)
#insert
names.insert(0,"Anne")

names.remove("Edwin")
print(names)

```

```python
names =["Ann","Joy","Fancy","Kevin","Liz"]

#loops
for name in names:
    print(name)

#Using range

for i in range(len(names)):
    print(names[i])

listcopy=names.copy()
print(listcopy)
```
